```python
# Volatility Decay study
import numpy as np
import matplotlib.pyplot as plt

time=np.arange(0,100,1.)
price=[] ; decayRate= 0.01 ;twox, threex=[],[]
startPrice=100. ; i = 0 ; startPrice2=100. ; startPrice3=100.
for day in time:
    i+=1
    if (i % 2) == 0:
        startPrice=startPrice*(1.+decayRate)
        startPrice2=startPrice2*(1.+2*decayRate)
        startPrice3=startPrice3*(1.+3*decayRate)
    else:
        startPrice=startPrice*(1.-decayRate)
        startPrice2=startPrice2*(1.-2*decayRate)
        startPrice3=startPrice3*(1.-3*decayRate)
    price.append(startPrice)
    twox.append(startPrice2)
    threex.append(startPrice3)

price=np.array(price) ; twox=np.array(twox) ; threex=np.array(threex)
plt.figure(1,[10,10])
plt.plot(time,price,'b.-',label='1x leverage')
plt.plot(time,twox,'r.-',label='2x leverage')
plt.plot(time,threex,'y.-',label='3x leverage')
plt.ylabel('%')
plt.xlabel('days')
plt.legend()
plt.show()
```
