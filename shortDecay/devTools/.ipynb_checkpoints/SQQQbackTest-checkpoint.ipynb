```python
""" Test stratagy staring from every data point and propagatin forwards to
the end. return index of 3% trades made, largest drawdown and duration of it
along with date index.Once have it calculate and store ave profit% YoY aswell.
[ [ startindexSim [indextrade1,....],[drawdown,Startindex,endIndex] ],.....]
"""
results=[] ; capital=10000 ; target = 0.97 ; trig=False
for i in range(len(dataSet)-1):
    startIndex=i+1
    trades=[]
    for j in range(startIndex,len(dataSet)):
        #loop to the end
        if j == startIndex:
            # first iteration, get starting value
            openTrade=dataSet[i][1]
            currentPrice=dataSet[j][1]
            counter1=i ; counter2=j
            DD=0.
        else:
            currentPrice=dataSet[j][1]
            if currentPrice<=openTrade*target:
                #sucsess, close,compund,reopen
                var=[i,[j],[DD,counter1,counter2]]  # ie [simStart,[compoundTrade],[draw,drawStart,drawEnd]]

                trades.append(var)
                #reset/update trackers
                openTrade=dataSet[j][1]
                counter1=j ; counter2=j
                DD=0.
            else:
                #calculate drawdown
                percent = 100.*(currentPrice/openTrade)-100.
                if percent >= DD:
                    DD=percent
                else:
                    pass
                counter2+=1
    results.append(trades)
```

```python
#Extract drawndowns from data and plot as histogram    
drawdowns=[]
for sim in results:
    for trade in sim:
        drawdowns.append(trade[2][0])
binWidths=np.arange(1,110)
plt.figure(1,[20,20])
plt.hist(drawdowns,bins=binWidths)
plt.xlim(0,150)
plt.xlabel('% drawdown')
plt.ylabel('count')
plt.show()
```
