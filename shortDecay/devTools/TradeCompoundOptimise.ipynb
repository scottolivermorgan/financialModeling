```python
import matplotlib.pyplot as plt
import numpy as np

time=np.arange(1,2555.,1.)
decay=0.0012
capital=10000. ; rolling=10000. ; linear=[]
for t in time:
    v=(capital*(1.+decay))-capital
    rolling+=v
    linear.append(round(rolling,2))

def calc(target):
    oneP=[] ; tP=[]
    count=1
    
    decay=0.0012 ; #target= 1.01
    capital=10000.
    spread=1-((16.66-16.63)/16.63)
    tC=target*capital
    for t in time:
        #c+=1
        #if c>=3:
        #    break
        var=round(capital*(1.+decay*count),2)
        #print('working =',var)
        #print('target=',target*capital)
        if var >= round(tC,2):
            oneP.append(var*spread)
            tP.append(t)
            tC=round(target*var,2)
            count=1
            capital=var*spread
            #print('new Target=', tC)
        else:
            count+=1
    return oneP,tP


halfP, thalfP = calc(1.005)
#print(halfP)
oneP, tP = calc(1.01)
twoP, t2P = calc(1.02)
threeP, t3P = calc(1.03)
fiveP, t5P = calc(1.05)
tenP, t10P = calc(1.1)


ifig=plt.figure(1,figsize=[20.,10.])
plt.plot(thalfP,halfP,'c-',label='0.5%')
plt.plot(tP,oneP,'b-',label='1%')
plt.plot(t2P,twoP,'g-',label='2%')
plt.plot(t3P,threeP,'y-',label='3%')
plt.plot(t5P,fiveP,'k-',label='5%')
plt.plot(t10P,tenP,'m-',label='10%') #halfP, thalfP
plt.plot(time,linear,'r-',label='Linear')
plt.xlabel('Days', fontsize=18) ; plt.ylabel('Capital', fontsize=18)
plt.xlim(0,2555)# ;  plt.ylim(6000.,14000)
plt.ylim(10000,176000)
plt.grid()
plt.legend(prop={'size': 20})
plt.show()
margins={'0.5%':(halfP[-1]/halfP[0])*100,
    '1%':(oneP[-1]/oneP[0])*100,
         '2%':(twoP[-1]/twoP[0])*100,
         '3%':(threeP[-1]/threeP[0])*100,
         '5%':(fiveP[-1]/fiveP[0])*100,
         '10%':(tenP[-1]/tenP[0])*100,
        }
print(margins)
```
